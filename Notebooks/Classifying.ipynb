```python
import os
from PyPDF2 import PdfReader

def get_pdf_page_count(pdf_path):
    try:
        reader = PdfReader(pdf_path)
        return len(reader.pages)
    except Exception as e:
        print(f"Erreur lors de la lecture du fichier {pdf_path}: {e}")
        return None

def write_pdf_info(directory):
    output_file = 'pdf_info.txt'
    with open(output_file, 'w') as out_file:
        for root, _, files in os.walk(directory):
            for file in files:
                if file.lower().endswith('.pdf'):
                    pdf_path = os.path.join(root, file)
                    page_count = get_pdf_page_count(pdf_path)
                    if page_count is not None:
                        out_file.write(f"{pdf_path}: {page_count} pages\n")
                        print(f"{pdf_path}: {page_count} pages")

directory = 'C:\\Users\\msi\\Desktop\\FiscaBot\\PDF Data'  # Remplacez par le chemin de votre répertoire
write_pdf_info(directory)
```
